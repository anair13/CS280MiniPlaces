```python
def plot_loss(step = 50, pref=""):
    train_loss = np.load("snapshot_solver_deeper/training_loss.npy")
    test_loss = np.load("snapshot_solver_deeper/test_acc.npy")
    plt.figure(figsize=(5,5))
    plt.plot(test_loss)
    plt.axis([0, len(test_loss), 0, 1])
    plt.figure(figsize=(5,5))
    plt.plot(train_loss)
    plt.axis([0, len(train_loss), 0, 7])
    plt.figure(figsize=(5,5))
    plt.plot(range(0, len(train_loss), step), test_loss)
    plt.plot(train_loss)
    plt.axis([0, len(train_loss), 0, 7])
    return train_loss, test_loss
```
